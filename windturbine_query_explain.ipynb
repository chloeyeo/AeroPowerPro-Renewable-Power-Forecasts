```python
# Setup the pandas
import pandas as pd
pd.set_option('display.max_rows', None)
```

```python
# CSV file location
attributes_fp = 'https://osuked.github.io/Power-Station-Dictionary/object_attrs/dictionary_attributes.csv'
```

```python
# Read CSV file from link.
df_attrs = pd.read_csv(attributes_fp)
```

```python
# Query and read only wind turbine.
df2 = df_attrs.query("value == 'WIND'")
```

```python
len(df2[['id', 'dictionary_id']])
```

```python
# Create Numpy array for BMUnit to ID.
bm_to_id = dict(df2[['id', 'dictionary_id']].values)
```

```python
# Get Id from BMUnit
id = bm_to_id.get('ACHRW-1')
id 
```

```python
# Check existance by using 'is None'
if id is None:
    print("There is not BMUnit in the CSV file")
else:
    print(id)
```

```python
# Use that Id to query the data in the CSV files
df3 = df_attrs.query(f"dictionary_id == {id}")
df3
```

```python
# Essential data in the CSV files
LATITUDE = 'Latitude'
LONGITUDE = 'Longitude'
PLANT_TYPE = 'Plant Type'
HUB_HEIGHT = 'Hub-Height'
TURBINE_CAPACITY = 'Turbine Capacity (MW)'
NUM_TURBINES = 'No. of Turbines'
X_COORDINATE = "X-coordinate"
Y_COORDINATE = 'Y-coordinate'
```

```python
# Latitude, Longitude, Plant_type, and Hub_height
float(df3.query(f"attribute == '{LATITUDE}'")[['value']].values[0])
```

```python
# Turbine Capacity (MW), No. of Turbines, X-coordinate, and Y-coordinate

if df3.query(f"attribute == '{X_COORDINATE}'")[['id', 'value']].values.size > 0:
    # if it exist then return the values
    print("It is exist in the CSV file")
else:
    # if it does not exist then return None
    print("It is not exist in the CSV")
```
